```python
import pickle
import json
import networkx as nx

data = pickle.load(open('../data_ontology_graph_policheck.pickle', 'rb'))

nelist = { 'nodes' : [], 'edges' : [] }

removeNodes = ['device identifier', 'non-pii', 'mac address', 'sim serial number', 'guid', 'imei', 'serial number', 'gsfid', 'android id', 'advertising identifier', 'network information', 'router ssid', 'application instal', 'photo metadata']

for n in data.nodes:
    if n in removeNodes:
        continue
    if n == 'geographical location':
        n = 'geographic location'
        
    if n == 'billing information':
        n = 'payment information'
        
    if n == 'country code':
        continue

    if n == 'registration information':
        continue
        
    nelist['nodes'].append(n)
    
for s,t in data.edges:
    if s in removeNodes or t in removeNodes:
        continue
        
    if t == 'geographical location':
        t = 'gps'
        
    if t == 'billing information':
        t = 'payment information'
 
    if s == 'billing information':
        s = 'payment information'
        
    if t == 'altitude':
        continue

    if t == 'registration information' or s == 'registration information':
        continue
        
    if s == 'device information' and t == 'email address':
        continue
        
    if s == 'information about you' and t == 'demographic information':
        continue

    if s == 'user information' and t == 'gps':
        continue

    if s == 'device information' and t == 'gps':
        continue
        
    if s == 'registration information' and t == 'contact information':
        continue

    if s == 'phone number' and t == 'country code':
        continue
    
    if s == 'country code' or t == 'country code':
        continue
    
    if s == 'account information' and t == 'billing information':
        continue
        
    if s == 'demographic information' and t == 'gps':
        t = "geographic location"

    if s == 'profile information' and t == 'gps':
        t = "geographic location"
        
    if s == 'account information' and t == 'gps':
        t = "geographic location"
        
    if s == 'postal address' and t =='state':
        continue
  
    if s == 'postal address' and t =='city':
        continue
        
    if s == 'pii' and t == 'identifier':
        continue
        
    nelist['edges'].append({ 'source' : s, 'target' : t })
    

nelist['nodes'].append('gps')
nelist['edges'].append({ 'source' : 'gps', 'target' : 'geographic location'})
nelist['edges'].append({ 'source' : 'gps', 'target' : 'altitude'})
nelist['nodes'].append('speed')
nelist['edges'].append({ 'source' : 'gps', 'target' : 'speed'})
nelist['nodes'].append('heading')
nelist['edges'].append({ 'source' : 'gps', 'target' : 'heading'})

nelist['nodes'].append('postal address')
nelist['edges'].append({ 'source' : 'geographic location', 'target' : 'postal address'})

nelist['nodes'].append('city')
#nelist['edges'].append({ 'source' : 'postal address', 'target' : 'city'})
nelist['edges'].append({ 'source' : 'postal address', 'target' : 'country'})
nelist['edges'].append({ 'source' : 'postal address', 'target' : 'zip code'})
#nelist['edges'].append({ 'source' : 'postal address', 'target' : 'state'})
nelist['edges'].append({ 'source' : 'zip code', 'target' : 'state'})
nelist['edges'].append({ 'source' : 'zip code', 'target' : 'city'})

#nelist['edges'].append({ 'source' : 'phone number', 'target' : 'country code'})


nelist['edges'].append({ 'source' : 'payment information', 'target' : 'postal address'})
nelist['edges'].append({ 'source' : 'payment information', 'target' : 'contact information'})
nelist['edges'].append({ 'source' : 'payment information', 'target' : 'postal address'})
nelist['edges'].append({ 'source' : 'shipping information', 'target' : 'postal address'})


    
with open('ontology.json', 'w') as outputfile:
    outputfile.write(json.dumps(nelist, indent=4))
    

g = nx.DiGraph()

for n in nelist['nodes']:
    g.add_node(n)
    
for edge in nelist['edges']:
    g.add_edge(edge['source'], edge['target'])

A = nx.drawing.nx_agraph.to_agraph(g)
A.draw('OUT_DATA_ONT.png', prog='dot', args='-Granksep=2.0')
```

```python
nelist = { 'nodes' : [], 'edges' : [] }
for n in g.nodes:
    nelist['nodes'].append(n)
    
for s,t in g.edges:
    nelist['edges'].append({ 'source' : s, 'target' : t })
    
with open('ontology.json', 'w') as outputfile:
    outputfile.write(json.dumps(nelist, indent=4))
    
```

```python
data = pickle.dump(g, open('../data_ontology_graph_policheck.pickle', 'wb'), protocol=2)
```
